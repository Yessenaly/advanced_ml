```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from scipy.stats import multivariate_normal
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('owid-covid-data.csv')
df = df.loc[df['location'] == 'Russia'].reset_index(drop=True)
df = df[['date', 'total_cases', 'new_cases']]
df['date'] = pd.to_datetime(df['date'])
```

```python
train_start = pd.Timestamp('2020-03-03')
train_end = train_start + pd.DateOffset(days=49)
df = df.loc[df['date'] >= train_start].reset_index(drop=True)
train = df.loc[(df['date'] <= train_end)]
test = df.loc[(df['date'] > train_end)]
```

```python
df_x = np.array(df.index)
df_y = np.array(df['total_cases'])
```

```python
train.set_index('date')[['total_cases']].plot(figsize=(20, 5))
```

```python
train['index'] = train.index
train['const'] = 1
```

```python
x = np.array(train[['const', 'index']])
y = np.array(train['total_cases'])
y = np.log(y)
```

## Линейная регрессия

```python
model = LinearRegression()
model.fit(x, y)
```

## Апостериорное распределение

```python
N = 4000
X = np.linspace(-2, 2, N)
Y = np.linspace(-2, 2, N)
X, Y = np.meshgrid(X, Y)
pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X
pos[:, :, 1] = Y
```

```python
cur_mu, cur_sigma = np.array([0, 0]), 2 * np.array([[1, 0], [0, 1]])
Z = multivariate_normal.pdf(pos, mean=cur_mu, cov=cur_sigma)
```

```python
def myplot_heatmap(Z):
    # Make the plot
    plt.axis('equal')
    plt.xlim((-2, 2))
    plt.ylim((-2, 2))
    plt.pcolormesh(X, Y, Z, cmap=plt.cm.jet)
    
def bayesian_update(mu, sigma, x, y, sigma_noise=.25):
    sigma_n = np.linalg.inv(np.linalg.inv(sigma) + (1 / sigma_noise) * np.transpose(x).dot(x))
    temp1 =  np.linalg.inv(cur_sigma).dot(cur_mu)
    temp2 = (1 / sigma_noise) * x.T.dot(y)
    mu_n =  sigma_n.dot(temp1 + temp2)
    return mu_n, sigma_n
```

```python
myplot_heatmap(Z)
```

```python
cur_mu, cur_sigma = bayesian_update(cur_mu, cur_sigma, x, y)
Z = multivariate_normal.pdf(pos, mean=cur_mu, cov=cur_sigma)
```

```python
plt.figure(figsize=(8, 8))
plt.xlim((1.5, 2))
plt.ylim((0.1, 0.3))
plt.pcolormesh(X, Y, Z, cmap=plt.cm.jet)
```

## Предсказание через сэмплирование

```python
def exponents_mean(x, lines):
    res = []
    for i in range(lines.shape[0]):
        cur = np.exp(lines[i][0] + lines[i][1] * x)
        res.append(cur)
    return np.array(res).mean(axis=0)
```

```python
def plot_sampled_lines(mu, sigma, n_lines, n_days):
    res = np.random.multivariate_normal(cur_mu, cur_sigma, n_lines)
    xs = np.linspace(0, n_days, 1000)
    plt.figure(figsize=(20, 10))
    for i in range(res.shape[0]):
        plt.plot(xs, np.exp(res[i][0] + res[i][1] * xs), 'black', linewidth=0.2)
    avg = exponents_mean(xs, res)
    plt.plot(xs, avg, 'red', linewidth=3.0)
    plt.plot(df_x[:n_days + 1], df_y[:n_days + 1], 'b', linewidth=2.0)
    return res
```

### Красная линия - предсказательное распределение

```python
sampled_lines = plot_sampled_lines(cur_mu, cur_sigma, n_lines=500, n_days=50)
```

```python
sampled_lines = plot_sampled_lines(cur_mu, cur_sigma, n_lines=500, n_days=100)
```

### Предсказания в точках

```python
# 1 мая
day_num = (pd.Timestamp('2020-05-01') - pd.Timestamp('2020-03-03')).days
exponents_mean(day_num, sampled_lines)
```
